```python
!pip install beautifulsoup4
!pip install requests
from bs4 import BeautifulSoup
import requests
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import collections
from bs4 import Comment
from io import StringIO
import sys 
```

```python
page_mlb = requests.get('https://www.baseball-reference.com/teams/BOS/2020.shtml')
soup = BeautifulSoup(page_mlb.text, 'html.parser')

table = []
for item in soup.find_all(text=lambda text:isinstance(text, Comment)):
    data = BeautifulSoup(item,"html.parser")
    for tr in data.find_all("tr"):
         table.append(tr.find_all(text=True))
        
table[1:10]



```

```python
player_data = pd.DataFrame.from_dict(table)
df = player_data.loc[1:16,player_data.columns[1:14]]
df
```

## Rename Columns...

```python
df.rename(columns={1:"Uni", 2: "Name", 3: "Flag", 4:"Nationality", 
                   5:"Position", 6:"OnActv", 7:"Age", 8:"Bats", 9:"Throws", 
                   10:"Height", 11:"Weight", 12:"DoB", 13:"1stYr"}, inplace=True)

df
```

## Converting Height from String to Centimeters...

```python
def parse_ht(ht):
    ht_ = ht.split("' ")
    ft_ = float(ht_[0])
    in_ = float(ht_[1].replace('"',''))*0.1
    return (30.48*ft_) + (30.48*in_)

df["Height"] = df["Height"].apply(lambda x:parse_ht(x))
df
```

## Drop extraneous data...

```python
df.drop(columns=['Flag', 'OnActv', 'Uni'], inplace=True)
df
```
